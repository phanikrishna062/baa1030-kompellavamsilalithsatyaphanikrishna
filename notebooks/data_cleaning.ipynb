## `Data Cleaning`

```python
df1_cleaned = unicef_indicator_df1[unicef_indicator_df1["time_period"] == 2016].copy()
```

```python
df1_cleaned = df1_cleaned[df1_cleaned["obs_value"].notnull()]
```

```python
df1_cleaned.rename(columns={
    "country": "Country",
    "indicator": "Indicator",
    "sex": "Gender",
    "obs_value": "Value"}, inplace=True)
```

```python
df1_cleaned.to_csv('../data/processed/unicef_indicator1_cleaned.csv', index=False)
```

```python
mno.bar(df1_cleaned)
```

---

```python
df2_cleaned = unicef_indicator_df2.drop(columns=[
    'unit_multiplier', 'observation_status', 'observation_confidentiality',
    'time_period_activity_related_to_when_the_data_are_collected'], errors='ignore')
```

```python
df2_cleaned.rename(columns={
    "country": "Country",
    "indicator": "Indicator",
    "sex": "Gender",
    "obs_value": "Value",
    "time_period": "Year"}, inplace=True)
```

```python
df2_cleaned = df2_cleaned.dropna(subset=["Value", "Indicator", "Country"])
```

```python
df2_cleaned.to_csv('../data/processed/unicef_indicator2_cleaned.csv', index=False)
```

```python
mno.bar(df2_cleaned)
```

---

```python
metadata_cleaned = metadata_df[metadata_df["year"] == 2016].copy()
```

```python
meta_cleaned = metadata_df[[
    "country", 
    "GDP per capita (constant 2015 US$)",
    "Birth rate, crude (per 1,000 people)",
    "Hospital beds (per 1,000 people)"]].dropna()
```

```python
meta_cleaned.rename(columns={"country": "Country"}, inplace=True)
```

```python
meta_cleaned.to_csv('../data/processed/unicef_metadata.csv', index=False)
```

```python
mno.bar(meta_cleaned)
```
